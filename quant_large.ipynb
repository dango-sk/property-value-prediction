#### T_PB&ROE

```python
import pandas as pd
from sklearn.linear_model import LinearRegression 
from scipy.stats import linregress
from sklearn.preprocessing import StandardScaler
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# 데이터 필터링 및 NaN 처리
sample = data[['T_ROE_Percent', 'T_PBR']]
sample.columns = ['X', 'y']

min_x, max_x = 0, 1
min_y, max_y = 0, 15

sample['X'] = np.where((sample['X'] >= min_x) & (sample['X'] < max_x), sample['X'], np.nan)
sample['y'] = np.where((sample['y'] >= min_y) & (sample['y'] < max_y), sample['y'], np.nan)

# NaN이 아닌 값들만 사용
filtered_sample = sample.dropna(subset=['X', 'y'])
X = filtered_sample[['X']]
y = filtered_sample[['y']]

# 선형 회귀 모델 훈련
model = LinearRegression()
model.fit(X, y)

# coef(기울기)와 intercept(절편) 추출
coef = model.coef_[0][0]
intercept = model.intercept_[0]

# 예측값 계산 (X에 대한 예측)
filtered_sample['pred_y'] = model.predict(X)

# R² 값 계산
r2 = r2_score(y, filtered_sample['pred_y'])

print(f"coef: {coef:.4f}")
print(f"intercept: {intercept:.4f}")
print(f"r2_score: {r2:.4f}")

# NaN을 0으로 채우기 (이 부분은 상황에 따라 변경 가능)
sample = sample.fillna(0)

# 전체 데이터에 FITTED_PBR 계산
sample['FITTED_PBR'] = coef * sample['X'] + intercept
sample['ROE_PBR_Combined'] = sample['FITTED_PBR'] / sample['y'] - 1

# 원본 데이터에 새로운 열 추가
data = pd.concat([data, sample[['FITTED_PBR', 'ROE_PBR_Combined']]], axis=1)

# nan 값 처리 
data['ROE_PBR_Combined'] = data['ROE_PBR_Combined'].replace([np.inf, -np.inf], np.nan)
```

```python
data_score = data[['Code', 'Name', 'T_PER', 'F_PER', 'T_EV/EBITDA', 'F_EV/EBITDA', 'T_PBR', 'F_PBR', 'T_P/CF', 
'ROE_PBR_Combined', 'EVIC_ROIC_Combined', 'PE_EPSG_Combined','EVEBIT_EBITG_Combined', 
'T_SPSG', 'F_SPSG', 'EPS_M', 'PRICE_M']]
```

```python
import numpy as np

def assign_score_low_over0(column):
    # NaN 값과 0 이하의 값을 제외한 후 퍼센타일 계산
    sorted_values = np.sort(column.dropna()[column.dropna() > 0])
    q1, q2, q3 = np.percentile(sorted_values, [25, 50, 75])

    # 점수 부여 함수
    def score(value):
        if np.isnan(value):
            return 0  # NaN인 경우 0 반환
        elif value <= 0:
            return 0  # 0 이하인 경우에도 0 반환
        elif value <= q1:
            return 4
        elif value <= q2:
            return 3
        elif value <= q3:
            return 2
        else:
            return 1
    
    # 점수 적용
    return column.apply(score)

def assign_score_low(column):
    # NaN 값을 제외한 후 퍼센타일 계산
    sorted_values = np.sort(column.dropna())
    q1, q2, q3 = np.percentile(sorted_values, [25, 50, 75])

    # 점수 부여 함수
    def score(value):
        if np.isnan(value):
            return 0  # NaN인 경우 그대로 반환
        elif value <= q1:
            return 4
        elif value <= q2:
            return 3
        elif value <= q3:
            return 2
        else:
            return 1
    
    # 점수 적용
    return column.apply(score)    

def assign_score_high(column):
    # NaN 값을 제외한 후 퍼센타일 계산
    sorted_values = np.sort(column.dropna())
    q1, q2, q3 = np.percentile(sorted_values, [25, 50, 75])
    
    # 점수 부여 함수
    def score(value):
        if np.isnan(value):
            return 0  # NaN인 경우 그대로 반환
        elif value >= q3:
            return 4
        elif value >= q2:
            return 3
        elif value >= q1:
            return 2
        else:
            return 1
    
    # 점수 적용
    return column.apply(score)
```

```python
low_score_columns_over0 = ['T_PER', 'F_PER', 'T_EV/EBITDA', 'F_EV/EBITDA', 'T_PBR', 'F_PBR', 'T_P/CF']
low_score_columns = ['PRICE_M']
high_score_columns = ['ROE_PBR_Combined', 'EVIC_ROIC_Combined', 'PE_EPSG_Combined', 'EVEBIT_EBITG_Combined'
, 'T_SPSG', 'F_SPSG', 'EPS_M']

for col in low_score_columns_over0:
    data_score[col + '_Score'] = assign_score_low_over0(data_score[col])
    
# 각 컬럼에 맞는 스코어링 함수 적용
for col in high_score_columns:
    data_score[col + '_Score'] = assign_score_high(data_score[col])
    
for col in low_score_columns:
    data_score[col + '_Score'] = assign_score_low(data_score[col])

result = data_score[['Name', 'Code']].join(data_score.filter(like='_Score'))

print(result.columns)
```

### Weighting

```python
df = pd.read_excel(file_path, sheet_name='Weight_Large', header=1)
weighting_values = df['Weighting'].dropna().values

score_columns = [col for col in result.columns if '_Score' in col]

for i, col in enumerate(score_columns):
    result[col] = result[col] * weighting_values[i]

result['Total_Score'] = result[score_columns].sum(axis=1)
result = result.sort_values(by='Total_Score', ascending=False)

data['MKT_CAP_Mil'] = data['MKT_CAP_Mil'].apply(lambda x: '{:.2f}'.format(x))
data['MKT_CAP_Mil'] = data['MKT_CAP_Mil'].astype(float)

result = pd.merge(result, data[['MKT_CAP_Mil']], left_index=True, right_index=True)
result['MKT_CAP_Rank'] = result['MKT_CAP_Mil'].rank(ascending=False, method='min').astype(int)

result['Total_Score_Rank'] = result['Total_Score'].rank(ascending=False, method='min').astype(int)
```
